```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# import data

# default parameters
train_score_defaultParams = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AutoML/train_score_defaultParams.csv')
train_test_score_defaultParams = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AutoML/train_test_score_defaultParams.csv')
crossvalidated_defaultParams = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AutoML/10fCV_defaultParams.csv')
# with automl optimized parameters
train_score_automl = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AutoML/train_score_AutoMLParams.csv')
train_test_score_automl = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AutoML/train__test_score_AutoMLParams.csv')
crossvalidated_automl = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AutoML/10fCV_AutoMLParams.csv')
```

```python
for i in range(15):
  x = [train_score_defaultParams.iloc[i][0] + str(' default'), train_score_automl.iloc[i][0] + str(' optimized')]
  h = [train_score_defaultParams.iloc[i][1], train_score_automl.iloc[i][1]]
  plt.figure(figsize=(14,5))
  plt.bar(x, h)
  plt.show()
```
